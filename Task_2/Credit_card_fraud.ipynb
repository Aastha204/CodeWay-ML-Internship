```python
# importing necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
#importing data files
data_test=pd.read_csv('fraudTest.csv')
data_train=pd.read_csv('fraudTrain.csv')
```

```python
#seeing train dataset
data_train.head()
```

```python
# Split the training data into features (X_train) and labels (y_train)
# target variable = is_fraud
X_train=data_train.drop('is_fraud',axis=1)
y_train=data_train['is_fraud']
```

```python
#Split the testing data into features (X_test) and labels (y_test)
X_test=data_test.drop('is_fraud',axis=1)
y_test=data_test['is_fraud']
```

```python
# Standardize the features
# scaler=StandardScaler()
# X_train=scaler.fit_transform(X_train)
# X_test=scaler.fit_transform(X_test)

# giving error due to non-numeric data
```

```python
# Exclude non-numeric columns from standardization for training data
numeric_columns=X_train.select_dtypes(include=['number']).columns
X_train_numeric=X_train[numeric_columns]
```

```python
# exclude non-numeric columns from standardization for testing data
numeric_columns_test=X_test.select_dtypes(include=['number']).columns
X_test_numeric=X_test[numeric_columns_test]
```

```python
# Now Standardize the feature
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train_numeric)
X_test_scaled=scaler.fit_transform(X_test_numeric)
```

```python
# Apply SMOTE to the training data
smote=SMOTE(random_state=42)
X_train_resampled,y_train_resampled=smote.fit_resample(X_train_scaled,y_train)
```

```python
# fine-tune Random Forest Hyperparameters 
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(RandomForestClassifier(class_weight='balanced', random_state=42), param_grid, cv=3, scoring='f1')
grid_search.fit(X_train_resampled, y_train_resampled)
best_rf_model = grid_search.best_estimator_
```

```python
# Access feature importances
feature_importances = best_rf_model.feature_importances_
```

```python
#Logistic Regression
logistic_model=LogisticRegression()
logistic_model.fit(X_train_scaled,y_train)
logistic_predictions=logistic_model.predict(X_test_scaled)
```

```python
#Decision Tree
tree_model=DecisionTreeClassifier()
tree_model.fit(X_train_scaled,y_train)
tree_predictions=tree_model.predict(X_test_scaled)
```

```python
# Random Forest
forest_model=RandomForestClassifier(n_estimators=100,class_weight='balanced',random_state=42)
forest_model.fit(X_train_resampled,y_train_resampled)
forest_predictions=forest_model.predict(X_test_scaled)
```

```python
# Evaluate model
def evaluate_model(predictions,model_name):
    accuracy=accuracy_score(y_test,predictions)
    precision=precision_score(y_test,predictions)
    recall=recall_score(y_test,predictions)
    f1=f1_score(y_test,predictions)
    
    print(f"-----{model_name}-----")
    print(f"Accuracy:{accuracy:.4f}")
    print(f"Precision:{precision:.4f}")
    print(f"Recall:{recall:.4f}")
    print(f"F1-score:{f1:.4f}")
    print("\n")
    
```

```python
# Evaluate Logistic Regression
evaluate_model(logistic_predictions,"Logistic Regression")
```

```python
# Evaluate Decision Tree
evaluate_model(tree_predictions,"Decision Tree")
```

```python
# evaluate Random Forest
evaluate_model(forest_predictions,"Random Forest")
```
